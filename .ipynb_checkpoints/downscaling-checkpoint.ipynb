```python
db = cym.dbopen("main.sqlite")
ev = cym.Evaluator(db=db, write=False)
```

```python
dfp = tm.get_power(ev)
dfp.plot('Time','Value',figsize = (20,12))
```

```python
dfp_La = tm.get_power(ev, facilities=['lwr_a'])
dfp_Lb = tm.get_power(ev, facilities=['lwr_b'])
dfp_Sa = tm.get_power(ev, facilities=['sfr_a'])
dfp_Sb = tm.get_power(ev, facilities=['sfr_b'])
```

```python
dfp_sc = dfp.copy()
dfp_sc['Value'] *= 1/10.
dfp_La_sc = dfp_La.copy()
dfp_La_sc['Value'] *= 1./10.
dfp_Lb_sc = dfp_Lb.copy()
dfp_Lb_sc['Value'] *= 1./10.
dfp_Sa_sc = dfp_Sa.copy()
dfp_Sa_sc['Value'] *= 1./10.
dfp_Sb_sc = dfp_Sb.copy()
dfp_Sb_sc['Value'] *= 1./10.


ax = dfp_sc.plot('Time','Value',figsize = (20,12))
dfp_La_sc.plot('Time','Value',figsize = (20,12), ax=ax)
dfp_Lb_sc.plot('Time','Value',figsize = (20,12), ax=ax)
dfp_Sa_sc.plot('Time','Value',figsize = (20,12), ax=ax)
dfp_Sb_sc.plot('Time','Value',figsize = (20,12), ax=ax)
```

```python
def intergize(df, fraction = 1):
    df_unit = pd.DataFrame()

    df_unit['Time'] = df['Time']
    df['Value'] *= 1/fraction
    df_unit['Unit'] = df['Value'].apply(lambda x: int(x) + 1 if x > int(x) else int(x))
    return df_unit


def build(df):
    df_build['Build'] = (df['Unit'] - df['Unit'].shift(1))
    return df_build
```

```python
dfp_La_sc_unit = intergize(dfp_La_sc, 900)
dfp_Lb_sc_unit = intergize(dfp_Lb_sc, 900)
dfp_Sa_sc_unit = intergize(dfp_Sa_sc, 360)
dfp_Sb_sc_unit = intergize(dfp_Sb_sc, 360)


```

```python
ax2 = dfp_La_sc_unit.plot('Time','Unit',figsize = (20,12))
dfp_Lb_sc_unit.plot('Time','Unit',figsize = (20,12), ax=ax2)
dfp_Sa_sc_unit.plot('Time','Unit',figsize = (20,12), ax=ax2)
dfp_Sb_sc_unit.plot('Time','Unit',figsize = (20,12), ax=ax2)
```
